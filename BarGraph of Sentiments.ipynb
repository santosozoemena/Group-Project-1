```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns


with open("city_sentiments.csv") as sentiments:
    avg_sentiment = pd.read_csv(sentiments)
    
avg_sentiment = avg_sentiment[["Miami",
                 "New York",
                 "Los Angeles",
                 "Philadelphia",
                 "Houston",
                 "Atlanta",
                 "Chicago"]]

avg_sentiment    
```

```python
target_cities = ["Miami",
                 "New York",
                 "Los Angeles",
                 "Philadelphia",
                 "Houston",
                 "Atlanta",
                 "Chicago"]


avg_list = []
for city in target_cities:
    avg_list.append(avg_sentiment[city].mean())

avg_list
```

```python
sns.set()
x_axis = np.arange(len(target_cities))
    
plt.figure(figsize = (25,10))    
    
    
sentiment_analysis = plt.bar(x_axis,
                             avg_list,
                             color = {"darkred",
                                      "darkorange",
                                      "gold",
                                      "black",
                                      "darkblue",
                                      "darkgreen",
                                      "red"})

tick_locations = [value for value in x_axis]
tick_labels = target_cities
plt.xticks(tick_locations, tick_labels )
plt.ylim(0.165,0.169)
plt.xlabel("Cities")
plt.ylabel("Avg Compound Sentiment")
plt.title("City Sentiments of Restaurants Based on Twitter")

plt.savefig("City Sentiments of Restaurants Based on Twitter.png")

plt.show()
```
