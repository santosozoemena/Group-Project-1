```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
with open("city_sentiments.csv") as sentiments:
    heat_df = pd.read_csv(sentiments)
    
heat_df = heat_df[['Restaurant Names',
                               'Chicago', 
                               'New York', 
                               'Houston',
                               'Philadelphia',
                               'Los Angeles',
                               'Atlanta',
                               'Miami']]
heat_df
```

```python
target_cities = ["Miami",
                 "New York",
                 "Los Angeles",
                 "Philadelphia",
                 "Houston",
                 "Atlanta",
                 "Chicago"]

city_array_info = []
for city in target_cities:
    city_array_info.append(heat_df[city])
    
```

```python
heatmap_array = np.array(city_array_info)
heatmap_array
```

```python
sentiment_heat = heat_df.pivot()
```

```python
sns.set()
plt.figure(figsize = (25,10)) 




ax = sns.heatmap(heatmap_array,
                 vmin=0.1, 
                 vmax = 0.4, 
                 center = 0)





ax.invert_yaxis()

plt.show()
```
