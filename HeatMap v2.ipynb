```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
with open("city_sentiments.csv") as sentiments:
    heat_df = pd.read_csv(sentiments)
    
heat_df = heat_df[['Restaurant Names',
                               'Chicago', 
                               'New York', 
                               'Houston',
                               'Philadelphia',
                               'Los Angeles',
                               'Atlanta',
                               'Miami']]
heat_df
```

```python
chicago_df = heat_df[["Restaurant Names", "Chicago"]]
chicago_df
```

```python
chicago_data = np.array(chicago_df["Chicago"])

ax = sns.heatmap(chicago_data)

ax.invert_yaxis()



plt.show()
```
