```python
coordinates = []
geo_format = []
for city in target_cities:
    location = geolocator.geocode(city)
    geo = ('%s,%s,20mi' %(location.latitude, location.longitude))
    geo_format.append(geo)
    coordinates.append({"Latitude": location.latitude,
                        "Longitude": location.longitude,
                        "Geo Format" : geo_format[0],
                        "City": city})
    
coordinates_df = pd.DataFrame.from_dict(coordinates)
coordinates_df = coordinates_df.set_index("City", inplace = False)
```

```python
heatmap_df = []
geo_coordinates = []
coordinates_data = pd.read_csv("city_coordinates_dataframe.csv")
sentiment_df = pd.read_csv("LA-restaurant.csv")


def heatmap(element): 
    for element in target_cities:
        for index, row in coordinates_data.iterrows():
            geo_coordinates.append(row["Geo Format"])



        for index, row in sentiment_df.iterrows():
            public_tweets = api.search(row["User Name"], 
                                       rpp=10, 
                                       result_type="recent", 
                                       geocode = geo_coordinates[0])

            for tweet in public_tweets["statuses"]:
                compound = analyzer.polarity_scores(tweet["text"])["compound"]

                heatmap_df.append({'Compound_Sentiment' : compound,
                                   'City' : element,
                                   'Time' : tweet["created_at"]})          

# heatmap("Chicago")    
# heatmap("New York")
# heatmap("Houston")
# heatmap("Philadelphia")   
# heatmap("Los Angeles")
# heatmap("Atlanta")
# heatmap("Miami")
    
    
for element in target_cities:
    heatmap(element)

raw_heat = pd.DataFrame.from_dict(heatmap_df)


    
```
